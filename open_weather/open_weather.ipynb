```python
import os # operating system functionalities
import requests
from dotenv import load_dotenv # hide sensitive data ie api key
```

```python
path_to_env = '.env'
load_dotenv(path_to_env)

# variable to store API key
API_KEY = os.getenv('API_KEY')
```

## Make First API Call

* Construct the url for the current weather data endpoint.

### Required parameters:

* `q`: City name eg., `q=London`
* `units`: Optional, Possible values, use `metric` cc Kenyan education system eg cm, mm, kgs, 
* `appid`: API key, credentials

```python
# Defined API URL and Parameters for Nairobi City
nai_url = f'http://api.openweathermap.org/data/2.5//weather?q=Nairobi&units=metric&lang=en&appid={API_KEY}'
```

```python
# make the API request 
response = requests.get(nai_url)
data = response.json()
```

```python
# preview the data
data
```

```python
# extract data 
city_name = data['name']
temperature = data['main']['temp']
description = data['weather'][0]['description']

print(f"The temperature in {city_name} is {temperature} degrees celcius. The weather description is {description}.")
```

## Make call for 10 Cities

```python
# define cities
cities = ['Nairobi', 'Kampala', 'Sydney', 'Paris', 'London', 'Tokyo', 'Kisumu', 'Tunis', 'Cairo', 'Berlin']

cities_weather_data = {}

for city in cities:
    url = f'http://api.openweathermap.org/data/2.5//weather?q={city}&units=metric&lang=en&appid={API_KEY}'
    response = requests.get(url)
    cities_weather_data[city] = response.json()

```
